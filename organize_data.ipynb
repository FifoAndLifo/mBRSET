```python
import pandas as pd
from organize_data import organize_images
```

### BRSET

```python
df = pd.read_csv('/opc/davidres/retina/mbrset/data/labels.csv')
df
```

```python
df['3_DR'] = df['DR_ICDR'].apply(lambda x: 'Normal' if x == 0 else ('Non-proliferative' if x in [1, 2, 3] else 'Proliferative'))
df['2_DR'] = df['DR_ICDR'].apply(lambda x: 'Normal' if x == 0 else 'Diabetic Retinopathy')
df['patient_sex'] = df['patient_sex'].apply(lambda x: 'Male' if x == 1 else 'Female')

df['image_id'] = df['image_id']+'.jpg'
df.head()
```

```python
image_column = 'image_id'
image_path = './BRSET/images/'
result_dir = './BRSET/'

organize_images(df, image_column=image_column, class_column='3_DR', image_path=image_path, result_dir=result_dir+'3_DR')
organize_images(df, image_column=image_column, class_column='2_DR', image_path=image_path, result_dir=result_dir+'2_DR')
organize_images(df, image_column=image_column, class_column='patient_sex', image_path=image_path, result_dir=result_dir+'patient_sex')
```

### mBRSET

```python
df = pd.read_csv('./mBRSET/labels.csv')
df
```

```python
df['3_DR'] = df['final_icdr'].apply(lambda x: 'Normal' if x == 0 else ('Non-proliferative' if x in [1, 2, 3] else 'Proliferative'))
df['2_DR'] = df['final_icdr'].apply(lambda x: 'Normal' if x == 0 else 'Diabetic Retinopathy')
df['sex'] = df['sex'].apply(lambda x: 'Male' if x == 1 else 'Female')

df.head()
```

```python
image_column = 'file'
image_path = './mBRSET/images/'
result_dir = './mBRSET/'

organize_images(df.dropna(subset=['final_icdr']), image_column=image_column, class_column='3_DR', image_path=image_path, result_dir=result_dir+'3_DR')
organize_images(df.dropna(subset=['final_icdr']), image_column=image_column, class_column='2_DR', image_path=image_path, result_dir=result_dir+'2_DR')
organize_images(df, image_column=image_column, class_column='sex', image_path=image_path, result_dir=result_dir+'sex')
```
